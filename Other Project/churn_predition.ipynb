```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
churn_data = pd.read_csv("WA_Fn-UseC_-Telco-Customer-Churn.csv")
churn_data
```

```python
churn_data.info()
```

```python
churn_data = churn_data.drop(["customerID", "gender", "StreamingMovies","SeniorCitizen"], axis = 1)
churn_data
```

```python
from sklearn.preprocessing import LabelEncoder
encoded_data = {}
list = ["Partner", "Dependents", "PhoneService", "MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup","DeviceProtection","TechSupport", "StreamingTV",	"Contract", "PaperlessBilling", "PaymentMethod", "Churn"]
for col in list:
 encoder = LabelEncoder()
 churn_data[col] = encoder.fit_transform(churn_data[col])
 encoded_data[col] = encoder
```

```python
churn_data
```

```python
churn_data = churn_data["TotalCharges"]
churn_data
```

```python
X = churn_data.drop(columns="Churn")
y = churn_data["Churn"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1000)
```

```python
churn_model = RandomForestRegressor()
churn_model.fit(X_train, y_train)
```

```python
accuracy = churn_model.score(X_test, y_test)
accuracy
```
